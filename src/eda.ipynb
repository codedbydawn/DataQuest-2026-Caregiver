```python
# removed WTBS_ columns since they are not relevant to our analysis
main_clean = main.drop(columns=[c for c in main.columns if c.startswith('WTBS_')])

print(main_clean.shape)

print([c for c in main_clean.columns if c.startswith('WTBS_')])
```

```python
for col in main_clean.columns:
    print(col)
```

```python
drop_exact = [
    'PUMFID',
    'WGHT_PER',
    'DV_PROXY',
    'EQFLAG',
    'SURVMNTH',
    'CXRFLAG',
    'DDEV_FL',
    'DDEX_FL'
]

drop_prefixes = ['WTBS_']
drop_module_prefixes = ['CAR', 'PRA', 'ARE', 'HAR', 'NPA', 'RPA', 'NFA', 'PG', 'AG']

cols_to_drop = []

for col in main_clean.columns:
    if col in drop_exact:
        cols_to_drop.append(col)
    elif any(col.startswith(prefix) for prefix in drop_prefixes):
        cols_to_drop.append(col)
    elif any(col.startswith(prefix) for prefix in drop_module_prefixes):
        cols_to_drop.append(col)

main_reduced = main_clean.drop(columns=cols_to_drop)

print("Dropped:", len(cols_to_drop))
print(cols_to_drop)
print(main_reduced.shape)

```

```python
#NOT SURE ABOUT THE CODE FROM HERE ON OUT

candidate_targets = [
    'ICS_40',
    'FIS_10A', 'FIS_10H',
]

existing_targets = [c for c in candidate_targets if c in main_reduced.columns]
print("Existing target columns:", existing_targets)
```

```python
import numpy as np
reserve_codes = [
    6, 9,
    96, 99,
    996, 999,
    9996, 9999,
    99996, 99999,
    999996, 999999
]

main_targets = main_reduced.copy()

for col in existing_targets:
    main_targets[col] = main_targets[col].replace(reserve_codes, np.nan)
```

```python
burnout_items = [
    'ICS_40',
    'FIS_10A', 'FIS_10B', 'FIS_10C', 'FIS_10D',
    'FIS_10E', 'FIS_10F', 'FIS_10G', 'FIS_10H',
    'CRH_10', 'CRH_20', 'CRH_30'
]

burnout_items = [c for c in burnout_items if c in main_targets.columns]

main_targets['burnout_score'] = main_targets[burnout_items].mean(axis=1)

print("Burnout items used:", burnout_items)
print(main_targets['burnout_score'].describe())
```

```python
main_targets = main_targets[main_targets['burnout_score'].notna()].copy()

median_score = main_targets['burnout_score'].median()
main_targets['burnout_high'] = (main_targets['burnout_score'] >= median_score).astype(int)

print("Median cutoff:", median_score)
print(main_targets['burnout_high'].value_counts())
```

```python
target_cols_to_remove = burnout_items + ['burnout_score', 'burnout_high']

X = main_targets.drop(columns=[c for c in target_cols_to_remove if c in main_targets.columns])
y_reg = main_targets['burnout_score']
y_clf = main_targets['burnout_high']

print("X shape:", X.shape)
print("y_reg shape:", y_reg.shape)
print("y_clf shape:", y_clf.shape)
```
