```python
import pandas as pd
import pyreadstat

# Load SAS files
episode, episode_meta = pyreadstat.read_sas7bdat('episode.sas7bdat')
main, main_meta = pyreadstat.read_sas7bdat('maindata.sas7bdat')
```

```python
print('=== episode.sas7bdat ===')
print(f'Shape: {episode.shape}')
print(f'\nColumn labels:')
for col, label in episode_meta.column_names_to_labels.items():
    print(f'  {col}: {label}')
print(f'\ndtypes:\n{episode.dtypes}')
print(f'\nNull counts:\n{episode.isnull().sum()}')
episode.head()
```

```python
print('=== maindata.sas7bdat ===')
print(f'Shape: {main.shape}')
print(f'\nColumn labels:')
for col, label in main_meta.column_names_to_labels.items():
    print(f'  {col}: {label}')
print(f'\ndtypes:\n{main.dtypes}')
print(f'\nNull counts:\n{main.isnull().sum()}')
main.head()
```

```python
# Summary stats for numeric columns
print('=== episode describe ===')
display(episode.describe())
print('\n=== maindata describe ===')
display(main.describe())
```
